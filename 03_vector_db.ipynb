```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

# 1. Recuperamos el texto
ruta_txt = '/content/drive/MyDrive/u/cd2/texto_banco_atlas.txt'
with open(ruta_txt, 'r', encoding='utf-8') as f:
    texto_completo = f.read()

# 2. Recreamos los chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200, length_function=len)
chunks = text_splitter.split_text(texto_completo)

# 3. Cargamos nuestro modelo ganador (Multilingüe)
print("⏳ Cargando modelo de Embeddings Multilingüe...")
embeddings_ganador = HuggingFaceEmbeddings(model_name="paraphrase-multilingual-MiniLM-L12-v2")

# 4. CREAMOS LA BASE DE DATOS VECTORIAL
print("⏳ Vectorizando e insertando en ChromaDB (esto puede tomar un minuto)...")
# El persist_directory guarda la base de datos en una carpeta física
ruta_chroma = '/content/drive/MyDrive/u/cd2/chroma_db_atlas'

base_vectorial = Chroma.from_texts(
    texts=chunks,
    embedding=embeddings_ganador,
    persist_directory=ruta_chroma
)

print("\n✅ ¡Nivel 3 Completado!")
print(f"Total de fragmentos guardados en la base de datos: {base_vectorial._collection.count()}")
```
